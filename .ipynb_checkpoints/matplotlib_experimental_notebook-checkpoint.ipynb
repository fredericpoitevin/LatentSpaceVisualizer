```python
dataset = datasets.load_digits()
```

```python
data = dataset['images']
labels = dataset['target']
```

```python
data.shape
```

```python
labels.shape
```

```python
n, h, w = data.shape
```

```python
pca = PCA(n_components=2)
```

```python
z = pca.fit_transform(data.reshape((n, h * w)))
```

```python
z.shape
```

```python
x = z[:, 0]
y = z[:, 1]
```

```python
%matplotlib notebook

fig = plt.figure(figsize=(8,3))
ax = fig.add_subplot(121)
ax.scatter(x, y, s=5, c=labels, alpha=0.6)

# debug text is used to display the index of the data point
debug_text = ax.text(1.1, 1.1,'', horizontalalignment='center', verticalalignment='center', transform = ax.transAxes)

#fig.canvas.mpl_connect('button_press_event', button_press_event)
fig.canvas.mpl_connect('motion_notify_event', button_press_event)
```
